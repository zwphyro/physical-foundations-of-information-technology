# Печатоющие устройства

## Лазерный принтер

![](./images/15-01.png)

Развертка луча лазера по всей длине барабана осуществляется за счет вращения многогранника.

![](./images/15-04.png)

Барабан покрыт диэлектрическим слоем, который можно поляризовать с помощью коронирующих электродов.  
Лазерный луч деполяризует участки барабана, к которым будет прилипать тонер.  
Для повтора цикла заряд снимает деполяризирующая лампа.

## Акустооптический эффект

![](./images/15-02.png)

На акустичечески-активную среду подается волна с помощью пьезопреобразоватея.  
Внутри среды образуется стоячая волна (образуются области повышенной оптической плотности).  
Стоячая волна по сути образует дифракционную решетку.  
В результате дифракции один из лучей проходит через ограниченную зону и луч воздействует на барабан.  
Аккустооптический эффект используется, так как это наиболее быстрый способ взаимодействия внешнего напряжения и лазерного луча.

## Струйтый принтер, пьезоэффект

![](./images/15-05.png)

![](./images/15-06.png)

При подаче напряжения на пьезопреобразователь он деформируется, из сопла происходит выброс капель краски.

![](./images/15-07.png)

Также выстреливание можно совершать с помощью нагрева.

![](./images/15-08.png)

Каплю краски необзодимо зарядить, после чего заряженная капля проходит через конденсатори и откланяется, меняя напряжение на прастинах конденсатора можно регулировать угол откланения капли.

## Сублимационный принтер

![](./images/15-09.png)

Сублимация - при нагреве вещество переходит из твердого в газообразное состояние, минуя жидкую фазу. Это позволяет облаку краски хорошо впитываться в нагретую бумагу.
