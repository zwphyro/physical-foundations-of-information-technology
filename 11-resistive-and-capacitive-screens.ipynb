## Элекроемкость и проводники

![](./images/11-03.png)

Силовая линия - линия, касательная к которой является вектором электростатического поля.

## Граничные условия

Силовые линии выходят и входят под прямым углом, эквипотенциали расположены перпендикулярно силовым линиям.  
В проводниках в электростатическом поле отрицительные заряды скапливаются на границе, куда входят силовые линии, положительные - откуда выходят.

## Силовые линии и эквипотенциали электростатического поля, принцип работы емкостных экранов

![](./images/11-04.png)

Емкостной экран состоит из резисторов.  
При внесении проводящего электрический ток предмето происходит искажение электрического поля, емкость конденсатора увеличивается.  

![](./images/11-05.png)

Через конденсаторы проходит переменный электрический ток, фиксация изменения емкости конденсатора происходит при уменьшении силяы тока на фиксирующем резисторе.
