# P-N переход

## Собственные полупроводники, n-полупроводники, p-полупроводники

![](./images/12-01.png)

Полупроводник в чистом виде без примесей называется собственным полупроводником.  
В чистом виде собственные полупроводники не проводят электрический ток, собственный полупроводник начинает проводить ток, если в него добавить примесь.  
Примеры собственных полупроводников - кремний и германий.

### Электроны и дырки

Для того, чтобы получить n-полупроводник, в собственный полупроводник необходимо добавить примесь, имеющую валентность больше, чем у собственного полупроводника (донор), в данном случае появляется свободный электрон.  
Для того, чтобы получить p-полупроводник, в собственный полупроводник необходимо добавить примесь, имеющую валентность меньше, чем у собственного полупроводника (акцептор), в двнном случае появляется дырка.
В n-полупроводнике основными носителями зарядов являются электроны, в p-полупроводниках - дырки.

## P-N переход, диоды

![](./images/12-02.png)

В месте контакта p- и n-полупроводников образуется обедненная область, не дающая электронам из n-части проникать в p-часть.

![](./images/12-03.png)

При прямом смещении размер обедненной области уменьшается, и электроны из n-пролупроводника проникают в p-полупроводник, тем самым обеспечивая течение тока через полупроводник.

![](./images/12-04.png)

При обратном смещении размер обедненной области увеличивается, и электроны не имеют возможности перемещаться из n-области в p-область, таким образом полупроводник препятствует прохождению тока.

## Диодный мост

![](./images/12-05.png)

Диоидный мост используется для выпрямления переменного тока. На вход диоидного моста приходит переменный ток, на выходе (с резистора) снимается пульсирующий сигнал, для его выпрямления используется конденсатор.
