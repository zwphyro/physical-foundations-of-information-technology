# Квантовые компьютеры. Квантовая телепортация.

## Современные проблемы вычислений

Увеличивать плотность расположения транзисторов в процессоре до бесконечности невозможно.  
Соременные компьютеры плохо справляются с задачами разложения числа на простые множители, решение задачи коммивояжера, построение объемной модели поверхности Земли.

## Кубит

![](./images/16-03.png)

Кубит обладает множеством устойчевых состояний, которые определяются вероятностными комплексными коэффициентами $A$ и $B$.  
Особенность кубита - как только происходит процесс измерения, кубит меняет свое состояние либо на 1, либо на 0.



## Физическая реализация кубитов

![](./images/16-06.png)

В охлажжденной области располагаются ионы, которые фиксируются на некотором расстоянии друг от друга с помощью стоячей волны.  
Управление/считывание состояний ионнов происходит с помощью лазерных лучей.

## Квантовая запутанность

**Квантовая запутанность** - явление, при котором квантовое состояние группы объектов является взаимосвязанным, что позволяет передовать информацию мгновенно.

![](./images/16-01.png)

Квантово запутаннымя являются электроны, находящиеся на орбитали в составе атома вещества.

![](./images/16-02.png)

Квантово запутанными являются поляризованные фотоны, прошедшие через кристалл с двойным лучеприломлением.
