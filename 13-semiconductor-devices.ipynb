# Полупроводниковые приборы

## Биполярные транзисторы (усилители, ключи, логические элементы)

![](./images/13-01.png)

При подаче напряжения на базу дырки проходят через прямое соединение в n-область, часть дырок становятся побочными носителями заряда, и проникают в правую p-часть.  
Соотношения:  
$ I_Э = I_Б + I_К $  
$ I_Б << I_Э $  
$ I_Э \approx I_К $

![](./images/13-02.png)

![](./images/13-03.png)

**Ключ**  
При подаче импульса на базу ключ "открывается" и лампочка начинает гореть.

**Усилитель**  
Ток на выходе (коллекторе) коррелирует с током базе, но его амплитуда усиливается за счет внешнего питания. Конденсатор позволяет запретить пропускание постоянного напряжения.  

![](./images/13-08.png)

Когда транзистор не пропускает ток, то весь ток с внешнего питания подается на выход.

## ЦАП, АЦП

![](./images/13-10.png)

С помощью операционного усилителя происходит сравнение сигнала на входе с деленным сигналом с напражения питания, далее с помщью шифратора выходы с операционных усилителей преобразуются в двоичное число.

![](./images/13-11.png)

На выходе регистра оказываются цифры преобразуемого числа, проходя через резисторы выходные значения "взвешиваются" (младший бит - наибольшее сопротивление), псле чего все сигналы суммируются и усиливаются с помощью операционного усилителя.

## Счетчики, дешифраторы

![](./images/13-12.png)

Счетчик реализуется с помощью триггеров, триггер изменяет свое значение при каждом отрицательном фронте входного сигнала, выход каждого триггера соединен со входом следующего триггера.

![](./images/13-13.png)

Дешифратор используется для преобразования одного кода в другой.
