# Магнитные носители информации

## Ферриты и ферромагнетики

![](./images/08-01.png)

$ \vec{J} $ - вектор намагниченности.  
Ферромагнетики характеризуются высокой намагниченностью, высокой магнитной восприимчивостью, высокой относительной магнитной проницаимостью.

## Процессы намагничевания феррита, гистерезис, точка Кюри

![](./images/08-02.png)

Внутреннее пространство ферромагнетика разделяется на домены со своей намагниченностью.  
При нулевом внешнем магнитном поле суммарная намагниченность равна 0.  

![](./images/08-03.png)

При уменьшении внешнего магнитного поля намагниченность остановиться в точке, называемой остаточной намагниченностью.  
При намагничивании вещества в противоположном магнитном поле, то в итоге будет достигнута точка насыщения в обратном направлении.  
Чтобы размагнитить материал, необходимо поместить материал во внешнее поле величиной $H_c$ и резко убрать это поле, или локально нагреть материал до температуры, называемой точкой Кюри $T_k$.

## Запись и чтение аналоговой и цифровой информации

![](./images/08-04.png)

В зависимости от внешнего магнитного поля, отстаточная намагниченность принимает различные значения.

![](./images/08-05.png)

Необходимо иметь возможность записать два состояния.  
Петля гистерезиса должна приближаться к прямоугольной форме.  

![](./images/08-06.png)

Слева - записывающая головка, справа - считывающая.  
Наконечник считывающей головки должен обладать свойством зависимости сопротивления от внешнего магнитного поля (магнитно-резистивный материал).  

![](./images/08-07.png)

Заряд, движущийся в некоторм магнитном поле будет отклонятсья.  
Заряды, положительные (справа) или отрицательные (слева), под действием магнитного поля собираются на одной из пластин, на противоположной пластине скапливается заряд с противоположным знаком, возникает напряжение между данными пластинами (наличие напряжения $\Rightarrow$ наличие магнитного поля).

![](./images/08-08.png)

Переменный магнитный поток создает ЭДС.  
Для считывания магнитного поля с помощью закона Фарадея необходимо знать, с какой скоростьюи в какус сторону движется считывающая головка.
